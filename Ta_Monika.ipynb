# **RELIABILITY**

```python
import numpy as np

# Definisikan indikator dan itemnya
indikator_items = {
    'Trust Stance': ['TS1', 'TS2', 'TS3', 'TS4'],
    'Integrity': ['IN1', 'IN2', 'IN3'],
    'Disclosure of Intentions': ['DoI1', 'DoI2', 'DoI3'],
    'Honesty': ['Hon1', 'Hon2', 'Hon3'],
    'Purchase/Visit Intention': ['PVI1', 'PVI2', 'PVI3'],
    'Loyalty': ['Loy1', 'Loy2', 'Loy3']
}

# Fungsi manual hitung Cronbach's Alpha
def cronbach_alpha(items_scores):
    items_scores = np.array(items_scores)
    item_variances = items_scores.var(axis=0, ddof=1)
    total_score_variance = items_scores.sum(axis=1).var(ddof=1)
    n_items = items_scores.shape[1]
    alpha = (n_items / (n_items - 1)) * (1 - item_variances.sum() / total_score_variance)
    return alpha

# Simpan hasil
hasil_reliabilitas = []

for indikator, items in indikator_items.items():
    if all(item in df.columns for item in items):
        skor_item = df[items]
        alpha = cronbach_alpha(skor_item)
        hasil_reliabilitas.append({
            'Indikator': indikator,
            'Item': ', '.join(items),
            'Cronbach\'s Alpha': round(alpha, 3),
            'Reliabel': 'Ya' if alpha >= 0.7 else 'Tidak'
        })

# Tabel hasil
tabel_reliabilitas = pd.DataFrame(hasil_reliabilitas)

# Tampilkan tabel
tabel_reliabilitas
```

# **Uji validitas**

```python
from scipy.stats import pearsonr

# Definisikan indikator dan itemnya
indikator_items = {
    'Trust Stance': ['TS1', 'TS2', 'TS3', 'TS4'],
    'Integrity': ['IN1', 'IN2', 'IN3'],
    'Disclosure of Intentions': ['DoI1', 'DoI2', 'DoI3'],
    'Honesty': ['Hon1', 'Hon2', 'Hon3'],
    'Purchase/Visit Intention': ['PVI1', 'PVI2', 'PVI3'],
    'Loyalty': ['Loy1', 'Loy2', 'Loy3']
}

# r-tabel untuk n = 264 dan α = 0.05
r_tabel = 0.120

# Simpan hasilnya di list
hasil_validitas = []

for indikator, items in indikator_items.items():
    # Buat skor total dari seluruh item indikator
    df[f'total_{indikator}'] = df[items].sum(axis=1)

    for item in items:
        df[f'total_lain'] = df[f'total_{indikator}'] - df[item]
        r, p = pearsonr(df[item], df['total_lain'])

        hasil_validitas.append({
            'Indikator': indikator,
            'Item': item,
            'r-hitung': round(r, 3),
            'Valid': 'Ya' if r >= r_tabel else 'Tidak'
        })

# Ubah ke DataFrame
tabel_validitas = pd.DataFrame(hasil_validitas)

# Tampilkan tabel hasil
tabel_validitas
```

# **Uji Normalitas**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kstest, probplot
import seaborn as sns

# 1. Baca file Excel
df = pd.read_excel('Data Base Monika.xlsx')

# 2. Tabel Hasil Uji Normalitas per Variabel (PERTAHANKAN INI)
item_columns = df.select_dtypes(include='number').columns
hasil_normalitas = []

for item in item_columns:
    data = df[item].dropna()
    if data.std() == 0 or len(data) < 2:
        continue

    data_norm = (data - data.mean()) / data.std()
    stat, p = kstest(data_norm, 'norm')
    hasil_normalitas.append({
        'Item': item,
        'Statistic': round(stat, 4),
        'p-value': round(p, 4),
        'Normal': 'Ya' if p >= 0.05 else 'Tidak'
    })

tabel_normalitas = pd.DataFrame(hasil_normalitas)
display(tabel_normalitas)  # Tampilkan tabel di awal

# 3. Gabungkan Semua Data Numerik untuk Plot
all_numeric_data = df.select_dtypes(include='number').values.flatten()
all_numeric_data = all_numeric_data[~np.isnan(all_numeric_data)]  # Hapus NaN
all_data_norm = (all_numeric_data - np.mean(all_numeric_data)) / np.std(all_numeric_data)

# 4. Buat Plot Gabungan
plt.figure(figsize=(14, 6))

# Histogram Gabungan
plt.subplot(1, 2, 1)
sns.histplot(all_data_norm, kde=True, stat='density', color='skyblue')
plt.title('Histogram Semua Variabel Numerik', fontweight='bold')
plt.xlabel('Nilai Terstandarisasi')
plt.ylabel('Densitas')

# P-P Plot Gabungan
plt.subplot(1, 2, 2)
probplot(all_data_norm, dist='norm', plot=plt)
plt.title('P-P Plot Semua Variabel Numerik', fontweight='bold')
plt.gca().get_lines()[0].set_markerfacecolor('blue')
plt.gca().get_lines()[0].set_markersize(5.0)

plt.tight_layout()
plt.show()

# 5. Uji Normalitas untuk Data Gabungan
stat_all, p_all = kstest(all_data_norm, 'norm')
print(f"\nHasil Uji Normalitas Data Gabungan:")
print(f"Statistic: {stat_all:.4f}, p-value: {p_all:.4f}")
print("Distribusi Normal" if p_all > 0.05 else "Bukan Distribusi Normal")
```

# **Uji Kruskal-Wallis Berdasarkan Kategori X**

```python
import pandas as pd
from scipy.stats import kruskal

# Ganti dengan path file Anda jika perlu
df = pd.read_excel("Data Base Monika.xlsx")

# Pastikan kolom kategori 'X' ada
kategori = df["X"].unique()
kategori.sort()

hasil_kruskal = []

# Uji Kruskal-Wallis untuk setiap variabel numerik berdasarkan kategori X
for kolom in df.select_dtypes(include='number').columns:
    if kolom == 'X':
        continue
    grup_data = [df[df['X'] == k][kolom].dropna() for k in kategori]
    if all(len(g) > 0 for g in grup_data):
        stat, p = kruskal(*grup_data)
        hasil_kruskal.append({
            'Variabel': kolom,
            'Statistic': round(stat, 4),
            'p-value': round(p, 4),
            'Signifikan': 'Ya' if p < 0.05 else 'Tidak'
        })

# Tampilkan hasil
hasil_df = pd.DataFrame(hasil_kruskal)
print("=== Hasil Uji Kruskal-Wallis Berdasarkan Kategori X ===")
print(hasil_df.to_string(index=False))
```

# **Uji Homogenitas Varians (Levene Test)**

```python
import pandas as pd
from scipy.stats import levene

# Baca file
df = pd.read_excel("Data Base Monika.xlsx")

# Inisialisasi
hasil_levene = []
kategori = df["X"].unique()
kategori.sort()

# Lakukan uji Levene untuk setiap kolom numerik terhadap kategori X
for kolom in df.select_dtypes(include='number').columns:
    if kolom == 'X':
        continue
    grup_data = [df[df['X'] == k][kolom].dropna() for k in kategori]
    if all(len(g) > 1 for g in grup_data):
        stat, p = levene(*grup_data)
        hasil_levene.append({
            'Variabel': kolom,
            'Statistic': round(stat, 4),
            'p-value': round(p, 4),
            'Homogen': 'Ya' if p > 0.05 else 'Tidak'
        })

# Tampilkan hasil
hasil_df = pd.DataFrame(hasil_levene)
print("=== Hasil Uji Homogenitas Varians (Levene Test) ===")
print(hasil_df.to_string(index=False))
```

# **Uji Homoskedastisitas (Breusch-Pagan)**

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# Baca ulang file Anda
df = pd.read_excel("Data Base Monika.xlsx")

# Gunakan model regresi: Trust ~ LevAnt + Transparancy
X = df[["X", "AvgTransparancy"]]
X = sm.add_constant(X)
y = df["AvgTrust"]

# Buat model regresi
model = sm.OLS(y, X).fit()

# Breusch-Pagan Test
bp_test = het_breuschpagan(model.resid, model.model.exog)
labels = ["LM Stat", "LM p-value", "F-Stat", "F p-value"]
hasil_bp = dict(zip(labels, bp_test))

# Tampilkan hasil
print("=== Uji Homoskedastisitas (Breusch-Pagan) ===")
for k, v in hasil_bp.items():
    print(f"{k}: {v:.4f}")

if hasil_bp["F p-value"] > 0.05:
    print("✅ Varians residual homogen (tidak ada masalah heteroskedastisitas).")
else:
    print("⚠️ Terdapat heteroskedastisitas (varians residual tidak homogen).")
```

# **Uji Kesesuaian Chi-Kuadrat**

```python
import pandas as pd
from scipy.stats import chisquare

# 1. Baca data
df = pd.read_excel("Data Base Monika.xlsx")

# 2. Hitung frekuensi aktual dari kategori 'X'
observed_freq = df['X'].value_counts().sort_index()
observed = observed_freq.values

# 3. Tentukan frekuensi harapan (uniform)
expected = [len(df['X']) / len(observed)] * len(observed)

# 4. Uji Chi-Square Goodness of Fit
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# 5. Tampilkan hasil
print("=== Uji Kesesuaian Chi-Kuadrat ===")
print(f"Chi-Square Statistic : {chi2_stat:.4f}")
print(f"P-Value              : {p_value:.4f}")

if p_value > 0.05:
    print("✅ Tidak ada perbedaan signifikan antara distribusi observasi dan harapan (sesuai).")
else:
    print("⚠️ Terdapat perbedaan signifikan antara distribusi observasi dan harapan (tidak sesuai).")

# 6. Tampilkan distribusi aktual vs harapan
print("\nFrekuensi Observasi:")
print(observed_freq)
print("\nFrekuensi Harapan (seragam):")
print(expected)
```

# **Uji Multikolinearitas**

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Load data
df = pd.read_excel("Data Base Monika.xlsx")

# Ambil variabel independen
X = df[['AvgTrust', 'AvgTransparancy', 'X']]

# Tambahkan konstanta untuk intercept
X_const = add_constant(X)

# Hitung VIF dan Tolerance
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i+1) for i in range(len(X.columns))]
vif_data["Tolerance"] = 1 / vif_data["VIF"]

# Cetak hasil
print(vif_data)
```

# **Uji Macro Hayes**

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# -------------------------------
# 1. Baca dan siapkan data
# -------------------------------
df = pd.read_excel('Data Base Monika.xlsx')

# Rename kolom agar sesuai model
df = df.rename(columns={
    "LevAnt": "X",                    # Predictor
    "AvgTrust": "M",                  # Mediator
    "AvgTransparancy": "W",          # Moderator
    "AvgCons.Behav": "Y"             # Outcome
})

# Mean-centering variabel X dan W
df["X_c"] = df["X"] - df["X"].mean()
df["W_c"] = df["W"] - df["W"].mean()

# Interaksi antara X dan W
df["XW"] = df["X_c"] * df["W_c"]

# -------------------------------
# 2. Estimasi model regresi
# -------------------------------

# Model 1: M ~ X + W + X*W (Moderasi)
model1 = smf.ols("M ~ X_c + W_c + XW", data=df).fit()

# Model 2: Y ~ X + M (Mediasi)
model2 = smf.ols("Y ~ X_c + M", data=df).fit()

# -------------------------------
# 3. Hitung efek tidak langsung (a*b)
# -------------------------------
a1 = model1.params["X_c"]
b = model2.params["M"]
indirect_effect = a1 * b

# -------------------------------
# 4. Bootstrapping untuk CI efek tidak langsung
# -------------------------------
n_boot = 1000
boot_indirects = []

for _ in range(n_boot):
    sample = df.sample(frac=1, replace=True)
    m1 = smf.ols("M ~ X_c + W_c + X_c:W_c", data=sample).fit()
    m2 = smf.ols("Y ~ X_c + M", data=sample).fit()
    boot_indirects.append(m1.params["X_c"] * m2.params["M"])

ci_lower = np.percentile(boot_indirects, 2.5)
ci_upper = np.percentile(boot_indirects, 97.5)

# -------------------------------
# 5. Tampilkan hasil
# -------------------------------
print("Model 1 Summary (M ~ X + W + X*W):")
print(model1.summary())

print("\nModel 2 Summary (Y ~ X + M):")
print(model2.summary())

print(f"\nIndirect Effect (X ➝ M ➝ Y): {indirect_effect:.4f}")
print(f"95% Confidence Interval (Bootstrap): [{ci_lower:.4f}, {ci_upper:.4f}]")

if ci_lower > 0 or ci_upper < 0:
    print("✅ Indirect effect is statistically significant.")
else:
    print("❌ Indirect effect is NOT statistically significant.")
```

# **UJI ANOVA**

```python
import pandas as pd
import scipy.stats as stats
from statsmodels.formula.api import ols
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file Excel
df = pd.read_excel('Data Base Monika.xlsx')

# Mengganti nama kolom 'AvgCons.Behav' agar tidak error
df = df.rename(columns={"AvgCons.Behav": "AvgConsBehav"})

# Pastikan kolom 'X' adalah numerik dan ubah ke kategori jika perlu
df['X'] = pd.to_numeric(df['X'], errors='coerce') # Coerce non-numeric to NaN
df['X'] = df['X'].astype('category') # Convert to categorical

## 1. ANOVA untuk AvgTrust berdasarkan X
print("\n" + "="*50)
print("ANOVA untuk AvgTrust berdasarkan X")
print("="*50)

model_trust = ols('AvgTrust ~ C(X)', data=df).fit()
anova_trust = sm.stats.anova_lm(model_trust, typ=2)

print("\nTabel ANOVA:")
print(anova_trust)

# Uji post-hoc Tukey jika ANOVA signifikan
if anova_trust['PR(>F)'][0] < 0.05:
    print("\nHasil signifikan (p < 0.05), melakukan uji post-hoc Tukey:")
    tukey_trust = pairwise_tukeyhsd(endog=df['AvgTrust'], groups=df['X'], alpha=0.05)
    print(tukey_trust)
else:
    print("\nTidak ada perbedaan signifikan (p > 0.05) antara kelompok")

# Visualisasi
plt.figure(figsize=(8, 6))
sns.boxplot(x='X', y='AvgTrust', data=df)
plt.title('Perbandingan AvgTrust berdasarkan Kategori X')
plt.xlabel('Kategori X')
plt.ylabel('AvgTrust')
plt.show()

## 2. ANOVA untuk AvgTransparancy berdasarkan X
print("\n" + "="*50)
print("ANOVA untuk AvgTransparancy berdasarkan X")
print("="*50)

model_trans = ols('AvgTransparancy ~ C(X)', data=df).fit()
anova_trans = sm.stats.anova_lm(model_trans, typ=2)

print("\nTabel ANOVA:")
print(anova_trans)

if anova_trans['PR(>F)'][0] < 0.05:
    print("\nHasil signifikan (p < 0.05), melakukan uji post-hoc Tukey:")
    tukey_trans = pairwise_tukeyhsd(endog=df['AvgTransparancy'], groups=df['X'], alpha=0.05)
    print(tukey_trans)
else:
    print("\nTidak ada perbedaan signifikan (p > 0.05) antara kelompok")

# Visualisasi
plt.figure(figsize=(8, 6))
sns.boxplot(x='X', y='AvgTransparancy', data=df)
plt.title('Perbandingan AvgTransparancy berdasarkan Kategori X')
plt.xlabel('Kategori X')
plt.ylabel('AvgTransparancy')
plt.show()

## 3. ANOVA untuk AvgConsBehav berdasarkan X
print("\n" + "="*50)
print("ANOVA untuk AvgConsBehav berdasarkan X")
print("="*50)

model_cons = ols('AvgConsBehav ~ C(X)', data=df).fit()
anova_cons = sm.stats.anova_lm(model_cons, typ=2)

print("\nTabel ANOVA:")
print(anova_cons)

if anova_cons['PR(>F)'][0] < 0.05:
    print("\nHasil signifikan (p < 0.05), melakukan uji post-hoc Tukey:")
    tukey_cons = pairwise_tukeyhsd(endog=df['AvgConsBehav'], groups=df['X'], alpha=0.05)
    print(tukey_cons)
else:
    print("\nTidak ada perbedaan signifikan (p > 0.05) antara kelompok")

# Visualisasi
plt.figure(figsize=(8, 6))
sns.boxplot(x='X', y='AvgConsBehav', data=df)
plt.title('Perbandingan AvgConsBehav berdasarkan Kategori X')
plt.xlabel('Kategori X')
plt.ylabel('AvgConsBehav')
plt.show()

## Uji Asumsi ANOVA
print("\n" + "="*50)
print("Uji Asumsi ANOVA")
print("="*50)

def test_anova_assumptions(data, group_col, value_col):
    groups = data[group_col].unique()

    # 1. Uji Normalitas (Shapiro-Wilk untuk setiap kelompok)
    print("\nUji Normalitas (Shapiro-Wilk):")
    for group in groups:
        group_data = data[data[group_col] == group][value_col]
        stat, p = stats.shapiro(group_data)
        print(f"{group}: stat={stat:.3f}, p={p:.3f}",
              "(Normal)" if p > 0.05 else "(Tidak Normal)")

    # 2. Uji Homogenitas Varians (Levene)
    group_data = [data[data[group_col] == g][value_col] for g in groups]
    stat, p = stats.levene(*group_data)
    print(f"\nUji Homogenitas Varians (Levene): stat={stat:.3f}, p={p:.3f}",
          "(Homogen)" if p > 0.05 else "(Tidak Homogen)")

# Uji asumsi untuk AvgTrust
print("\nAsumsi untuk AvgTrust:")
test_anova_assumptions(df, 'X', 'AvgTrust') # Ganti 'X_category' dengan 'X'

# Uji asumsi untuk AvgTransparancy
print("\nAsumsi untuk AvgTransparancy:")
test_anova_assumptions(df, 'X', 'AvgTransparancy') # Ganti 'X_category' dengan 'X'

# Uji asumsi untuk AvgConsBehav
print("\nAsumsi untuk AvgConsBehav:")
test_anova_assumptions(df, 'X', 'AvgConsBehav') # Ganti 'X_category' dengan 'X'
```
